Visualizing The Gender Gap In College Degrees
====================

A DataQuest Guided Project
---------------------

In this project, I aim to visualize the gender gap in college degrees. To compare multiple plots at once, I used subplots and customized the plots to reduce clutter. The US Department of Education released a dataset containing the percentage of  bachelor's degrees granted to women from 1970 to 2012. The dataset I am using below was cleansed by Randal Olson, a data scientist at University of Pennsylvania.

### Comparing Across All Degrees
Degrees have been categorized into 3 groups, displayed in the rows of the visual:
* STEM: Science, Technology, Engineering and Mathematics
* Liberal Arts
* Other

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(18, 9))


## Generate first row of line charts. STEM degrees.
for sp in range(0,6):
    ax = fig.add_subplot(3,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3, linewidth=2)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2002, 87, 'Women')
        ax.text(2005, 8, 'Men')
    elif sp == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2001, 32, 'Women')
        
## Generate second row of line charts. Liberal arts degrees.
for sp in range(6,11):
    counter = sp-6
    ax = fig.add_subplot(3,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[counter]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[counter]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3, linewidth=2)
    ax.set_title(lib_arts_cats[counter])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if counter == 0:
        ax.text(2002, 83, 'Women')
        ax.text(2005, 10, 'Men')

## Generate third column of line charts. Other degrees.
for sp in range(12,18):
    counter = sp-12
    ax = fig.add_subplot(3,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[counter]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[counter]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3, linewidth=2)
    ax.set_title(other_cats[counter])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if counter == 0:
        ax.text(2002, 70, 'Women')
        ax.text(2005, 25, 'Men')
        
    elif counter == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2001, 29, 'Women')

fig.subplots_adjust(hspace=1.0)

#For exporting as an image
#fig.savefig("gender_degrees.png")

plt.show()
```

### Conclusion
The visualization created allows us to quickly view the gender divide across each category and within each degree. It was interesting to note the degrees with the largest divides between men and women: Engineering being male dominated and majors such as Psychology, Foreign Languages, Health Professions, Public Administration and Education being female dominated. It was also interesting to observe the trends within those majors, for example, the % of women has been increasing in Engineering over time.
