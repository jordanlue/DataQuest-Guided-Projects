Customizing Data Visualizations
====================

A DataQuest Guided Project

In this project, I aim to recreate plots that are generated using high-level Pandas DataFrame plotting methods using lower-level Matplotlib methods. This project will allow me to gain more familiarity with customizing visualizations. 

### Introduction
The dataset has been loaded below.


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
recent_grads.describe()
```

```python
print(recent_grads.shape)
recent_grads = recent_grads.dropna()
print(recent_grads.shape)
```

### Scatter Plot from Pandas

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[["ShareWomen","Unemployment_rate"]])
```

### Scatter Plot Recreated

```python
fig = plt.figure(figsize=(6,4))
topleft = fig.add_subplot(2,2,1)
topright = fig.add_subplot(2,2,2)
bottomleft = fig.add_subplot(2,2,3)
bottomright = fig.add_subplot(2,2,4)

topleft.hist(recent_grads['ShareWomen'])
topright.scatter(recent_grads['Unemployment_rate'],recent_grads['ShareWomen'])
bottomleft.scatter(recent_grads['ShareWomen'],recent_grads['Unemployment_rate'])
bottomright.hist(recent_grads['Unemployment_rate'])

topleft.xaxis.set_visible(False)
topright.xaxis.set_visible(False)
topright.yaxis.set_visible(False)

fig.subplots_adjust(wspace=0, hspace=0)

topleft.set(ylim=(0,30),ylabel="ShareWomen")
topright.set(xlim=(0.0,0.2),ylim=(0.0,1.0))
bottomleft.set(xlim=(0.0, 1.0), ylim=(0.0,0.20),xlabel="ShareWomen", ylabel="Unemployment_rate")
bottomright.set(xlim=(0.0,0.20), xlabel="Unemployment_rate")

topleft.set_yticklabels([0, 5, 10, 15, 20, 25, 30])
bottomleft.set_yticklabels([0.00, 0.05, 0.10, 0.15])
bottomleft.set_xticklabels([0.0, 0.2, 0.4, 0.6, 0.8], rotation=90)
bottomright.set_xticklabels([0.00, 0.05, 0.10, 0.15], rotation=90)
```

```python
### Grouped Bar Chart from Pandas
```

```python
recent_grads["ShareMen"] = recent_grads["Men"]/recent_grads["Total"] 
arts = recent_grads[recent_grads["Major_category"] == "Arts"]
arts.set_index("Major", inplace=True)
arts[["ShareMen", "ShareWomen"]].plot(kind="bar")
```

### Grouped Bar Chart Recreated

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,1,1)

import numpy as np

locs = np.arange(0,arts.index.nunique())
w=0.35
bar_1 = ax.bar(left=locs, height=(arts["ShareMen"].tolist()), width=w)
ax.set_xticklabels(arts.index.tolist(), rotation=90)

offset_locs = locs+w
bar_2 = ax.bar(left=offset_locs, height=(arts["ShareWomen"].tolist()), width=w, color="green")
ax.set_xticks(offset_locs)
plt.legend((bar_1, bar_2), ("ShareMen", "ShareWomen"), loc="upper left")
plt.grid()
```

### Conclusion
Where possible, the built in high-level methods from Pandas and Seaborn are great and big time savers. With more practice and practical applications, I'll know when to work with the high-level methods, or when I need to customize using Matplotlib's lower-level methods directly.
